```python
def denoise_image(image):
    # Gaussian Blur (Removes random noise)
    blurred = cv2.GaussianBlur(image, (5, 5), 0)

    # Bilateral Filtering (Preserves edges while removing noise)
    denoised = cv2.bilateralFilter(blurred, 9, 75, 75)
    return denoised
```

```python
import cv2
import os
from tqdm import tqdm  # Progress bar

# Define input and output folders
INPUT_FOLDER = "/content/input_imag"    # Folder containing original images
OUTPUT_FOLDER = "/content/preprocessed_images"  # Folder to save processed images
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# Define preprocessing function
def preprocess_image(image_path, output_path):
    # Read Image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Skipping {image_path}, not a valid image!")
        return

    # Convert to Grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Denoising (Bilateral Filter)
    denoised = cv2.bilateralFilter(gray, 9, 75, 75)

    # Edge Detection (Canny)
    edges = cv2.Canny(denoised, 50, 150)

    # Resize Image to 256x256
    resized = cv2.resize(edges, (256, 256))

    # Save processed image
    cv2.imwrite(output_path, resized)

# Process all images in the folder
print(f"Processing {len(os.listdir(INPUT_FOLDER))} images...")

for file_name in tqdm(os.listdir(INPUT_FOLDER)):
    if file_name.endswith(('.jpg', '.jpeg', '.png')):  # Process only images
        input_path = os.path.join(INPUT_FOLDER, file_name)
        output_path = os.path.join(OUTPUT_FOLDER, file_name)
        preprocess_image(input_path, output_path)

print(f"\n✅ Preprocessing Complete! Processed images saved in: {OUTPUT_FOLDER}")
```

```python
import os
import shutil
import random

# Define paths
INPUT_FOLDER = "/content/preprocessed_images"
TRAIN_FOLDER = "/content/data/train"
VAL_FOLDER = "/content/data/val"
TEST_FOLDER = "/content/data/test"

# Create directories
for folder in [TRAIN_FOLDER, VAL_FOLDER, TEST_FOLDER]:
    os.makedirs(folder, exist_ok=True)

# Get all image filenames
all_images = os.listdir(INPUT_FOLDER)
random.shuffle(all_images)  # Shuffle to ensure randomness

# Split images into train (70%), val (20%), test (10%)
train_split = int(0.7 * len(all_images))
val_split = int(0.9 * len(all_images))

train_images = all_images[:train_split]
val_images = all_images[train_split:val_split]
test_images = all_images[val_split:]

# Move files to respective folders
for img in train_images:
    shutil.move(os.path.join(INPUT_FOLDER, img), os.path.join(TRAIN_FOLDER, img))

for img in val_images:
    shutil.move(os.path.join(INPUT_FOLDER, img), os.path.join(VAL_FOLDER, img))

for img in test_images:
    shutil.move(os.path.join(INPUT_FOLDER, img), os.path.join(TEST_FOLDER, img))

print("✅ Dataset split complete!")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Define CNN Model
model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(256, 256, 1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # Crack or No Crack
])

# Compile Model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Summary of Model
model.summary()
```

```python
from tensorflow.keras import layers, Model

def unet_model(input_size=(256, 256, 1)):
    inputs = layers.Input(input_size)

    # Encoder
    conv1 = layers.Conv2D(64, (3,3), activation='relu', padding='same')(inputs)
    conv1 = layers.Conv2D(64, (3,3), activation='relu', padding='same')(conv1)
    pool1 = layers.MaxPooling2D(pool_size=(2,2))(conv1)

    # Decoder
    up1 = layers.UpSampling2D(size=(2,2))(pool1)
    conv2 = layers.Conv2D(64, (3,3), activation='relu', padding='same')(up1)
    outputs = layers.Conv2D(1, (1,1), activation='sigmoid')(conv2)

    return Model(inputs, outputs)

# Compile U-Net Model
unet = unet_model()
unet.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
unet.summary()
```
